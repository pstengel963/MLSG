```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
# Set-up.
n = 10
xrng = np.random.default_rng(seed=41)
x = xrng.random(n)*n
dyrng = np.random.default_rng(seed=99)
dy = dyrng.normal(loc=0,scale=n/10.,size=n)
y = x + dy
w = np.dot(x,y)/np.dot(x,x)
xfit = np.linspace(x.min(), x.max(), 200)
yfit = w*xfit
wvar = np.linspace(0.5, 1.5, 20)
mse = np.zeros(20)
for i in range(20):
    mse[i] = np.sum(np.power(wvar[i]*x-y,2))
```

```python
plt.scatter(x,y)
plt.plot(xfit,yfit)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(wvar,mse)
plt.xlabel("w")
plt.ylabel("MSE")
plt.show()
```
